# Convertir Columnas de CSV a Json

Para poder utilizar las respuestas con la librer√≠a dataset, y a su vez en transformers

```python
import pandas as pd
import json
from pprint import pprint
from sklearn.model_selection import train_test_split
from shortuuid import uuid
```

```python
def JSON_samples(csv, col_resp, col_labels, nombre):
    """
    Crea los splits de entrenamiento y prueba en forma JSON

    Args:
        -csv: archivo csv leido por pandas
        -col_resp: numero de la columna de respuesta
        -col_labels: numero de lal columna de labels
        -nombre: nombre de la pregunta
    """

    x = csv.iloc[:,col_resp].tolist() #respuestas
    y =  csv.iloc[:,col_labels].tolist() #labels
    y = [True if i == 3 else False for i in y]
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42) #creamos los splits
    out = { #creamos el formato tipo json para cada split
        'train' : {'data' : [{'answer' : x_train[i], 'label' : y_train[i], 'uq' : uuid()} for i in range(len(x_train))]},
        'test' : {'data' : [{'answer' : x_test[i], 'label' : y_test[i], 'uq' : uuid()} for i in range(len(x_test))]}
    }

    #exportamos los splits como json

    with open(f'{nombre}_train.json', 'w') as fp:
        json.dump(out['train'], fp)
    
    with open(f'{nombre}_test.json', 'w') as fp:
        json.dump(out['test'], fp)
```

Leemos el csv

```python
url = 'https://github.com/ValverDEV/NLPCogEval/blob/main/palabras_corregidas_secundaria.csv?raw=true'
csv = pd.read_csv(url, encoding='latin1')
```

Ejecutamos para las columnas correspondientes

```python
JSON_samples(csv, 0, 1, 'pelota_gritos')
JSON_samples(csv, 2,3, 'pelota_pelos')
JSON_samples(csv, 4, 5, 'pelota_plata')
```
