# Convertir Columnas de CSV a Json

Para poder utilizar las respuestas con la librería dataset, y a su vez en transformers

```python
import pandas as pd
import json
from pprint import pprint
```

Importamos el dataset de un repositorio

```python
url = 'https://github.com/ValverDEV/NLPCogEval/blob/main/palabras_corregidas_secundaria.csv?raw=true'
csv = pd.read_csv(url, encoding='latin1')
```

Cargamos para la pregunta "explica lo que quiere decir pelota de plata" las respuestas y los label. En este caso hemos convertido los label a binarios (siendo 3 = 1, y los demás = 0)

```python
respuestas = csv.iloc[:,4].tolist()
labels = [True if i == 3 else False for i in csv.iloc[:,5].tolist()]
```

Unimos las respuetas y sus labels en un diccionario para convertirlo en JSON

```python
out = {}
out['datos'] = [{'answer' : respuestas[i], 'label' : labels[i]} for i in range(len(respuestas))]
```

## Exportamos nuestros datos en un archivo JSON

Luego los leemos para ver que haya funcionado correctamente

```python
with open('pelota_plata.json', 'w', encoding='latin1') as fp:
  json.dump(out, fp)
```

```python
with open('pelota_plata.json', 'r', encoding='latin1') as fp:
  data = json.load(fp)
```
