```python
import pandas as pd 
import datasets
from sklearn.model_selection import train_test_split
from shortuuid import uuid
```

```python
csv = pd.read_csv('palabras_corregidas_secundaria.csv', encoding='latin1')
```

```python
def csv_split(csv, col_resp, col_labels, nombre):
    ids = [uuid() for i in range(len(csv))]

    x = csv.iloc[:,col_resp].tolist()
    y = csv.iloc[:,col_labels].tolist()
    y = [True if i == 3 else False for i in y]
    id_train, id_test, x_train, x_test, y_train, y_test = train_test_split(ids, x, y, test_size=0.25, random_state=42)
    out = {
        'train' : pd.DataFrame({
            'id' : id_train,
            'answer' : x_train,
            'label' : y_train
        }),
        'test' :  pd.DataFrame({
            'id' : id_test,
            'answer' : x_test,
            'label' : y_test
        })
    }

    out['train'].to_csv(f'{nombre}_train.csv', index=False)
    out['test'].to_csv(f'{nombre}_test.csv', index=False)
```

```python
csv_split(csv, 0, 1, 'pelota_gritos')
csv_split(csv, 2, 3, 'pelota_pelos')
csv_split(csv, 4, 5, 'pelota_plata')
```
